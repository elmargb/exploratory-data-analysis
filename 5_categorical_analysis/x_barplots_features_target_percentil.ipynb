### examples

```python
# Crear un DataFrame de ejemplo
data = {'A': [1, 1, 1, 0, 0, 0],
        'B': [1, 1, 0, 0, 1, 1],
        'C': [1, 0, 0, 1, 1, 1]}

df = pd.DataFrame(data)

# Calcular la tabla cruzada
cross_tab = pd.crosstab(df['A'], [df['B'], df['C']], rownames=['A'], colnames=['B', 'C'])

# Crear un gráfico con Plotly Express
fig = px.imshow(cross_tab, labels=dict(x="B", y="C", color="Count"),
                x=cross_tab.columns.levels[1], y=cross_tab.columns.levels[2])

# Configurar el diseño del gráfico
fig.update_layout(title="Tabla Cruzada",
                  xaxis_title="B",
                  yaxis_title="C")

# Mostrar el gráfico
fig.show()
```

```python
# otra idea hacer cross tab de percentiles pero que luego haya funciones de agregacion de la media, std, por ejemplo
import pandas as pd
import numpy as np

# Crear un DataFrame de ejemplo con variables continuas
np.random.seed(42)
data = {'Variable1': np.random.randn(100),
        'Variable2': np.random.randn(100),
        'Variable3': np.random.randn(100)}

df = pd.DataFrame(data)

# Dividir las variables en percentiles
percentiles = [0, 25, 50, 75, 100]
df_percentiles = pd.DataFrame()

for col in df.columns:
    df_percentiles[col] = pd.cut(df[col], percentiles, labels=False)

# Calcular la media de cada grupo en percentiles
result_mean = df.groupby(df_percentiles.values.tolist()).mean()

# Calcular la desviación estándar de cada grupo en percentiles
result_std = df.groupby(df_percentiles.values.tolist()).std()

# Mostrar los resultados
print("Media por grupo en percentiles:")
print(result_mean)

print("\nDesviación estándar por grupo en percentiles:")
print(result_std)
```
